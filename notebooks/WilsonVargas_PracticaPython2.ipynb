# Ejercicio 16: DataFrame de estudiantes

**Enunciado:** crear un DataFrame con 10 estudiantes que incluya nombre, carrera, edad, promedio y ciudad.

**Lógica:** se utiliza pandas para crear una tabla de estudiantes. Luego se muestran las primeras filas, la información general, el promedio general, los estudiantes con promedio mayor o igual a 80 y finalmente se ordenan de mayor a menor promedio.

```python
import pandas as pd

datos = {
    "Nombre": ["Ana", "Luis", "Carlos", "María", "José", "Sofía", "Pedro", "Lucía", "Juan", "Gabriela"],
    "Carrera": ["Sistemas", "Derecho", "Medicina", "Sistemas", "Administración", "Psicología", "Civil", "Arquitectura", "Contaduría", "Industrial"],
    "Edad": [20, 21, 22, 19, 23, 20, 24, 22, 21, 20],
    "Promedio": [85, 78, 90, 82, 75, 95, 70, 88, 80, 92],
    "Ciudad": ["Tegucigalpa", "Comayagua", "La Ceiba", "Danlí", "Choluteca", "SPS", "Juticalpa", "Siguatepeque", "Tela", "Tegucigalpa"]
}

df = pd.DataFrame(datos)

print("Primeras 5 filas:")
display(df.head())

print("Información general:")
df.info()

promedio_general = df["Promedio"].mean()
print(f"\nPromedio general: {promedio_general:.2f}")

print("\nEstudiantes con promedio mayor o igual a 80:")
display(df[df["Promedio"] >= 80])

print("\nEstudiantes ordenados de mayor a menor promedio:")
display(df.sort_values(by="Promedio", ascending=False))
```

## Ejercicio 17: Análisis de ventas con pandas

**Enunciado:** crear un DataFrame de ventas con al menos 15 registros e identificar métricas clave del negocio.

**Lógica:** se construye un DataFrame con las columnas `fecha`, `vendedor`, `producto`, `categoría`, `cantidad` y `precio unitario`. Se calcula la columna `TotalVenta` multiplicando cantidad por precio, luego se agrupan los datos para obtener el total por vendedor, el producto más vendido, la categoría con mayor ingreso y se ordenan las ventas de mayor a menor.

```python
import pandas as pd

datos = {
    "fecha": [
        "2024-03-05","2024-03-12","2024-03-21",
        "2024-04-02","2024-04-18","2024-04-27",
        "2024-05-07","2024-05-14","2024-05-29",
        "2024-06-03","2024-06-19","2024-06-25",
        "2024-07-08","2024-07-17","2024-07-30"
    ],
    "vendedor": [
        "Ana","Ana","Ana",
        "Luis","Luis","Luis",
        "Carlos","Carlos","Carlos",
        "María","María","María",
        "Jorge","Jorge","Jorge"
    ],
    "producto": [
        "Laptop","Audífonos Bluetooth","Silla Ergonómica",
        "Monitor 27 pulgadas","Teclado Mecánico","Webcam HD",
        "Smartphone","Tablet","Cargador Inalámbrico",
        "Impresora","Mouse Inalámbrico","Escritorio",
        "Laptop Gamer","Micrófono USB","Lámpara LED"
    ],
    "categoria": [
        "Electrónica","Audio","Mobiliario",
        "Electrónica","Accesorios","Accesorios",
        "Telefonía","Telefonía","Accesorios",
        "Oficina","Accesorios","Mobiliario",
        "Electrónica","Audio","Iluminación"
    ],
    "cantidad": [2, 3, 1, 2, 4, 5, 3, 2, 6, 1, 7, 1, 2, 4, 5],
    "precio_unitario": [
        1500, 80, 350,
        300, 95, 75,
        850, 450, 35,
        220, 30, 480,
        2200, 120, 40
    ]
}

df = pd.DataFrame(datos)

df["TotalVenta"] = df["cantidad"] * df["precio_unitario"]

print("DataFrame completo")
print(df)

print("\nTotal vendido por vendedor")
print(df.groupby("vendedor")["TotalVenta"].sum())

print("\nProducto más vendido")
print(df.groupby("producto")["cantidad"].sum().idxmax())

print("\nCategoría con mayor ingreso")
print(df.groupby("categoria")["TotalVenta"].sum().idxmax())

print("\nVentas ordenadas de mayor a menor")
print(df[["producto", "precio_unitario", "TotalVenta"]].sort_values("TotalVenta", ascending=False))
```

## Ejercicio 18: Limpieza de datos con pandas

**Enunciado:** realizar un proceso básico de limpieza de datos utilizando pandas.

**Lógica:** se crea un DataFrame de clientes, se cuentan los valores nulos, se rellenan las ciudades vacías con un valor por defecto y se eliminan las filas que no tengan correo registrado.

```python
import pandas as pd
import numpy as np


datos = {
    "nombre": ["johnny", "Katherine", "Carlos", "María", "Marco"],
    
    "edad": [25, 30, np.nan, 28, 35],
    
    "correo": [
        "johnny@gmail.com",
        None,
        "carlos@gmail.com",
        "maria@gmail.com",
        None
    ],
    
    "ciudad": [
        "Tegucigalpa",
        "",
        "San Pedro Sula",
        "",
        "La Ceiba"
    ],
    
    "telefono": [
        "9999-1111",
        "8888-2222",
        None,
        "7777-3333",
        "6666-4444"
    ]
}

df = pd.DataFrame(datos)


print("DataFrame original:")
print(df)


print("\nValores nulos:")
print(df.isnull())


print("\nCantidad de valores nulos por columna:")
print(df.isnull().sum())


df["ciudad"] = df["ciudad"].replace("", "No especificada")


df_limpio = df.dropna(subset=["correo"])


print("\nDataFrame limpio:")
print(df_limpio)
```

## Ejercicio 19: Análisis de préstamos

**Enunciado:** analizar varias solicitudes de préstamo para conocer su estado, los montos solicitados y características generales de los clientes.

**Lógica:** se crea un DataFrame con las solicitudes de préstamo, se exploran los datos para identificar montos, estados y características clave de cada cliente.

Se utilizaron varios metodos de pandas para trabajar con los datos. **DataFrame()** permitio crear y organizar la tabla con la informacion de prestamos. **mean()** se uso para obtener el promedio de los montos solicitados. **value_counts** ayudo a contar cuantas solicitudes habia en cada estado. **groupby()** permitio agrupar los datos segun el estado de la solicitud, y junto con **mean()** sirivio para calcular el promedio del monto solicitado en cada grupo. Ademas, los filtros facilitaron mostrar solo ciertos registros, como los prestamos aprobados o los montos mayores a 50,000.

```python
import pandas as pd

data = {
    "cliente": ["Jorge", "Luis", "Carlos", "Maria", "Sofia", "Pedro"],
    "monto_sol": [45000, 60000, 30000, 80000, 52000, 40000],
    "plazo": [12, 24, 6, 36, 18, 12],
    "ingreso_men": [15000, 22000, 12000, 30000, 25000, 18000],
    "estado": ["Aprobado", "Pendiente", "Rechazado", "Aprobado", "Aprobado", "Pendiente"]
}

df = pd.DataFrame(data)

print("DataFrame de solicitudes de prestamo:")
print(df)

prom_monto = df["monto_sol"].mean()
print("\nMonto promedio solicitado:", prom_monto)

conteo_estados = df["estado"].value_counts()
print("\nCantidad de solicitudes por estado:")
print(conteo_estados)

aprobados = df[df["estado"] == "Aprobado"]
print("\nPrestamos aprobados:")
print(aprobados)

mayores = df[df["monto_sol"] > 50000]
print("\nClientes con monto solicitado mayor a 50,000:")
print(mayores)

promedio_est = df.groupby("estado")["monto_sol"].mean()
print("\nPromedio del monto solicitado por estado:")
print(promedio_est)
```

## Ejercicio 20: Mini análisis de datos tipo empresa

Se utilizaron varios metodos de pandas para trabajar con los datos de los empleados. DataFrame() permitio crear y organizar la informacion en forma de tabla. head() se uso para mostrar los primeros registros del DataFrame. describe() ayudo a obtener estadisticas generales de los salarios, como promedio, minimo y maximo. groupby() junto con mean() permitio calcular el salario promedio por departamento.

```python
import pandas as pd

data = {
    "nombre": ["Ana", "Carlos", "Maria", "Luis", "Sofia", "Pedro"],
    "departamento": ["IT", "Ventas", "RRHH", "IT", "Marketing", "Ventas"],
    "cargo": ["Programador", "Vendedor", "Reclutadora", "Analista", "Diseñadora", "Supervisor"],
    "salario": [25000, 18000, 22000, 30000, 21000, 27000],
    "años_exp": [2, 6, 4, 8, 3, 10],
    "ciudad": ["Tegucigalpa", "San Pedro", "La Ceiba", "Tegucigalpa", "Choluteca", "San Pedro"]
}

df = pd.DataFrame(data)

print("DataFrame de empleados:")
print(df)

print("\nPrimeros registros:")
print(df.head())

estadisticas = df["salario"].describe()
print("\nEstadisticas generales de salarios:")
print(estadisticas)

promedio_dep = df.groupby("departamento")["salario"].mean()
print("\nSalario promedio por departamento:")
print(promedio_dep)

mayor_salario = df[df["salario"] == df["salario"].max()]
print("\nEmpleado con salario mas alto:")
print(mayor_salario)

mayor_exp = df[df["años_exp"] > 5]
print("\nEmpleados con mas de 5 años de experiencia:")
print(mayor_exp)

ordenados = df.sort_values(by="salario", ascending=False)
print("\nEmpleados ordenados por salario:")
print(ordenados)

def nivel_experiencia(años):
    if años <= 2:
        return "Junior"
    elif años <= 5:
        return "Intermedio"
    else:
        return "Senior"

df["NivelExperiencia"] = df["años_exp"].apply(nivel_experiencia)

print("\nDataFrame con nivel de experiencia:")
print(df)
```
