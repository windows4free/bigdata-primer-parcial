## Respuesta a las 4 preguntas

**1. ¿Qué datos genera?**

Spotify genera datos de forma continua desde múltiples fuentes. Su infraestructura de entrega de eventos ingresa hasta 8 millones de eventos por segundo, con una media de 500.000 millones de eventos al día, lo que equivale alrededor de  70 TB de datos comprimidos diarios. Entre los datos que se registran están los eventos de escucha, metadatos de audio, datos contextuales y datos de podcasts. La plataforma registra internamente más de 600 tipos de datos distintos y gestiona más de 4.000 millones de listas de reproducción creadas por usuarios.

**2. ¿Quién los usa?**

Los datos son consumidos por varios actores. La propia empresa los usa a través de más de 220 proyectos activos de machine learning para alimentar sus sistemas de recomendación y tomar decisiones de producto. Los artistas y sellos discográficos acceden a analíticas de sus oyentes a través de Spotify for Artists para orientar decisiones de promoción y giras. Los anunciantes usan los datos para segmentar sus campañas según el contexto del oyente: estado de ánimo inferido, dispositivo y momento del día.

**3. ¿Qué problema resuelve?**

El principal valor que aporta el análisis de datos es la personalización de la experiencia musical. El sistema de recomendación facilita más de 16.000 millones de descubrimientos de artistas al mes, conectando a cada usuario con música que probablemente no conocía pero que encaja con sus gustos. Esto reduce la tasa de cancelación de suscripciones, mejora la experiencia de los usuarios en plan gratuito al hacer los anuncios más relevantes, y permite identificar artistas emergentes antes de que alcancen popularidad masiva, beneficiando a músicos independientes y al ecosistema musical en general.

**4. ¿Por qué no bastaría Excel?**

Excel tiene un límite de poco más de un millón de filas, una cantidad que Spotify supera en cuestión de segundos. Las bases de datos relacionales tradicionales tampoco escalan horizontalmente de forma eficiente a decenas de terabytes distribuidos. Además, Excel es una herramienta de análisis estático y no puede procesar flujos de datos en tiempo real, que es lo que requiere la personalización inmediata. Por otro lado, los datos de Spotify no son solo tablas: incluyen señales de audio analizadas con redes neuronales, texto procesado y grafos de relaciones entre usuarios, formatos que Excel no puede manejar. Finalmente, los algoritmos de recomendación que combinan filtrado colaborativo, análisis de audio y procesamiento de lenguaje natural requieren frameworks distribuidos como Apache Spark y clústeres en la nube, algo completamente fuera del alcance de una hoja de cálculo.


## Justificación de por qué se considera Big Data

El caso de Spotify cumple con las dimensiones que definen el Big Data. En cuanto al volumen, genera cientos de terabytes de datos comprimidos al día procedentes de millones de usuarios simultáneos. En cuanto a la velocidad, los eventos se procesan en tiempo real a una tasa de hasta 8 millones por segundo, algo que exige infraestructura de streaming distribuido. En cuanto a la variedad, los datos son estructurados (tablas de eventos), semiestructurados (metadatos de canciones) y no estructurados (señales de audio, texto de reseñas, grafos sociales). En cuanto a la veracidad, la empresa necesita garantizar la calidad de sus datos mediante lo que internamente llama "golden datasets", conjuntos de datos certificados disponibles para todos sus equipos de ingeniería. No se trata simplemente de una empresa con muchos datos: es un sistema donde el volumen, la velocidad y la variedad hacen imposible el uso de herramientas convencionales y donde el valor del negocio depende directamente de la capacidad de procesarlos.
